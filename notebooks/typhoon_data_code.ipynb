```python
# Import relevant python modules
import numpy as np
import pandas as pd
import glob
```

```python
# 1. Grab the paths of all CSV files in your folder
path = './data/typhoon-info' 
all_files = glob.glob(path + "/*.csv")

# 2. Use a list comprehension to read them all at once
li = [pd.read_csv(filename) for filename in all_files]

# 3. Concatenate them into one master DataFrame
df = pd.concat(li, axis=0, ignore_index=True)
```

```python
unique_locations = df['Location'].unique()

# fmt='%s' ensures the locations are saved as strings
np.savetxt('unique_locations.txt', unique_locations, fmt='%s')
```

```python
# Load your mapping
mapping_df = pd.read_csv('data/geospatial-data/location_mapping.csv')

# Merge it into your main dataframe
# This effectively 'tags' every row with its Region
df = df.merge(mapping_df, on='Location', how='left')



```
